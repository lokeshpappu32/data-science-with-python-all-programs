```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.rand(10)
y=np.random.rand(10)
z=np.random.rand(10)
colors=np.random.rand(10)
plt.scatter(x,y,s=z*1000,c=colors,alpha=0.7)
plt.show()
```

```python
plt.plot(x,y,c="r",mec="yellow",marker="o",linewidth=4,ms=20)
```
